```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
#display(df)

#This is the data we will use for this assignment
url = 'https://raw.githubusercontent.com/HamedTabkhi/Intro-to-ML/b4cf4ac187d647a443dc011c4ee85fbb42e26096/Dataset/D3.csv'
df = pd.read_csv(url)
print(df.head())
```

```python
#This will get all the input(X1,X2,X3) and output(Y) values into an array list
X1 = df.values[:,0]
X2 = df.values[:,1]
X3 = df.values[:,2]
Y = df.values[:,3]

#Training examples
m = len(X1)
n = len(X2)
o = len(X3)
p = len(Y)

print('X1 = ', X1[: m])
#print('X2 = ', X2[: n])
#print('X3 = ', X3[: o])
print('\n')
print('m = ', m)
```

```python
def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)  # Number of training examples
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        predictions = X.dot(theta)
        errors = np.subtract(predictions, y)
        sum_delta = (alpha / m) * X.transpose().dot(errors)
        theta -= sum_delta
        cost_history[i] = compute_cost(X, y, theta)

    return theta, cost_history

def compute_cost(X, y, theta):
    predictions = X.dot(theta)
    errors = np.subtract(predictions, y)
    sqrErrors = np.square(errors)
    J = 1 / (2 * m) * np.sum(sqrErrors)
    return J
```

```python
#X1
X_0 = np.ones((m,1))
X_1 = X1.reshape(m, 1)
X = np.hstack((X_0, X_1))
theta = np.zeros(2)
theta = [0., 0.]
iterations = 1500
alpha = 0.1
theta, cost_history = gradient_descent(X, Y, theta, alpha, iterations)
print('X1:')
print('Value of Theta = ', theta)
print('Value of Cost = ', cost_history)
print(f"Linear model: Y = {theta[0]} + {theta[1]}*X1")

#Data Plot for X1
plt.scatter(X[:, 1], Y, color='red', marker='*', label='Training Data')
plt.plot(X[:, 1], X.dot(theta), color='black', label='Linear Regression')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('X1')
plt.ylabel('Y')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```

```python
#Loss over iteration plot for X1
plt.plot(range(1, iterations + 1), cost_history, color='red')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('Number of iterations')
plt.ylabel('Cost (J)')
plt.title('Convergence of gradient descent')
plt.show()
```
